# working with dictionary

```python
student ={
    "name" : "bibek",
    "age": 20,
    "faculty":"BCE"
}
```

```python
## acccess val
print(student["name"])
```

```python
a=student.get("age")
print(a)
```

```python

student["age"]="21"
print(student)
```

```python
student["address"]="ktm"
print(student)
```

```python
del student["address"]
print(student)
```

```python
student.pop("age")
print(student)

```

```python
student.popitem()
print(student)
```

```python
student ={
    "name" : "bibek",
    "age": 20,
    "faculty":"BCE"
}
print(student)
```

```python
student.clear()
print(student)
```
