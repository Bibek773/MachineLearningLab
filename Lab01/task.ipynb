## Experiment 3

```python
# 3*3 matrix
matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
print(matrix)
```

```python
#diagolnal elements
for i in range(len(matrix)):
    print(matrix[i][i])
```

```python
matrix2 = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
result = []
for i in range(len(matrix)):
    row = []
    for j in range(len(matrix[0])):
        row.append(matrix[i][j] + matrix2[i][j])
    result.append(row)

print(result)
```

```python
matrix = [[1, 2, 3],
          [4, 5, 6],
          [7, 8, 9]]

matrix2 = [[1, 2, 3],
           [4, 5, 6],
           [7, 8, 9]]

result = []

for i in range(len(matrix)):
    row = []
    for j in range(len(matrix2[0])):
        total = 0
        for k in range(len(matrix2)):
            total += matrix[i][k] * matrix2[k][j]
        row.append(total)
    result.append(row)

print(result)
```
