# Working with pandas dataframe

```python
import pandas as pd
a = { 'Name': ['Ram', 'Shyam', 'Hari', 'Sita', 'Gita', 'bibek'],
    'Age': [24, 27, 22, 32, 20, 22],
    'City': ['Kathmandu', 'Bhaktapur', 'Bajhang', 'Lalitpur', 'Biratnagar', 'ilam'],
    'Marks': [14, 77, 62, 32, 80, 92]
}
```

```python
df = pd.DataFrame(a)
```

```python
df.to_csv('my_dataframe.csv', index=False)
print('DataFrame saved to my_dataframe.csv')
```

```python

data1 = {
    'Name': ['Ram', 'Shyam', 'Hari'],
    'Age': [24, 27, 22],
    'City': ['Kathmandu', 'Bhaktapur', 'Lalitpur']
}
data2 = {
    'Name': ['Sita', 'Gita', 'Rita'],
    'Age': [21, 21, 22],
    'City': ['Bajhang', 'Bajura', 'Doti']
}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

print('<-----------df1----------->')
print(df1)
print('<-----------df2----------->')
print(df2)

concatenated_df = pd.concat([df1, df2])
print('<------------concatenated df---------->')
print(concatenated_df)
     
```

```python
import openpyxl
# 1. Create the student data
student_data = {
    'Name': ['Ram', 'Shyam', 'Sita', 'Gita'],
    'Age': [21, 22, 20, 21],
    'Marks': [85, 92, 78, 95],
    'Grade': ['A', 'A+', 'B', 'A+']
}

# 2. Convert to a DataFrame
df = pd.DataFrame(student_data)

# 3. Export to an Excel (.xlsx) file
# Note: index=False prevents Pandas from adding a row-number column
df.to_excel('students_report.xlsx', index=False)

print("Excel file '.xlsx' created successfully!")
```
